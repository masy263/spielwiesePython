# TestBench Interpolation

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.style.use('dark_background')

y1 = np.array([0, 1, 4, 9, 16, 25])
x1 = np.arange(0, len(y1), 1)
x2 = np.arange(0, len(y1) - 1 + 0.1, 0.1)
y2 = np.interp(x2, x1, y1)

plt.plot(x1, y1, '-')
plt.plot(x2, y2, 'o')

plt.show()

print(x1)
print(y1)
print(x2)
print(y2)
```
